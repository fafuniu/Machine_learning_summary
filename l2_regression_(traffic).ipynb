```python
import numpy as np
import pandas as pd
from sklearn.linear_model import Ridge
from sklearn import cross_validation
import matplotlib.pyplot as plt
from sklearn.preprocessing import PolynomialFeatures
from sklearn import linear_model
data=pd.read_csv('regression.csv')
data=data.as_matrix()
print(type(data.shape))

plt.plot(data[:,5])

plt.show()
X=data[:,:5]
y=data[:,5]
print(X.shape)

#Generate polynomial and interaction features
poly=PolynomialFeatures(6)
X=poly.fit_transform(X)
print(X.shape)

train_set_x, test_set_x,train_set_y, test_set_y =cross_validation.train_test_split(X,y,test_size=0.3,random_state=0)

clf=Ridge(alpha=1.0,fit_intercept=True)
clf.fit(train_set_x,train_set_y)

linear=linear_model.LinearRegression()
linear.fit(train_set_x,train_set_y)

print(clf.score(test_set_x,test_set_y))

start=100
end=600

y_pre=clf.predict(X)
y_pre_L=linear.predict(X)
time=np.arange(start,end)
plt.plot(time,y[start:end],'b',label='real')
plt.plot(time,y_pre_L[start:end],'y',label='predict_Linear')
plt.plot(time,y_pre[start:end],'r',label='predict')

plt.legend(loc='upper left')
plt.show()
```
