```python
import matplotlib.pyplot as plt
import pandas as pd
from sklearn import linear_model
import numpy as np

data=pd.read_table("prices.txt",delimiter=",",header=None)
print(type(data))
#plt.scatter(data[0],data[1])
#plt.show()
datasets_X=data[0]
datasets_Y=data[1]
length=len(datasets_X)
print(type(datasets_X))
datasets_X=np.array(datasets_X).reshape([length,1])
datasets_Y=np.array(datasets_Y)
print(type(datasets_X))

minX=min(datasets_X)
maxX=max(datasets_X)
X=np.arange(minX,maxX).reshape([-1,1])

linear=linear_model.LinearRegression()
linear.fit(datasets_X,datasets_Y)


print("coefficients",linear.coef_)
print("intercept",linear.intercept_)

plt.scatter(datasets_X,datasets_Y,color='r')
plt.plot(X,linear.predict(X),c='b')
plt.xlabel("Area")
plt.ylabel("Price")
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd
from sklearn import linear_model
from sklearn.preprocessing import PolynomialFeatures
import numpy as np

data=pd.read_table("prices.txt",delimiter=",",header=None)
print(type(data))
#plt.scatter(data[0],data[1])
#plt.show()
datasets_X=data[0]
datasets_Y=data[1]
length=len(datasets_X)
print(type(datasets_X))
datasets_X=np.array(datasets_X).reshape([length,1])
datasets_Y=np.array(datasets_Y)
print(type(datasets_X))

minX=min(datasets_X)
maxX=max(datasets_X)
X=np.arange(minX,maxX).reshape([-1,1])

#poly_regression

poly_reg=PolynomialFeatures(degree=2)
X_poly=poly_reg.fit_transform(datasets_X)

print(X_poly.shape)

linear=linear_model.LinearRegression()
linear.fit(X_poly,datasets_Y)


print("coefficients",linear.coef_)
print("intercept",linear.intercept_)

plt.scatter(datasets_X,datasets_Y,color='r')
plt.plot(X,linear.predict(poly_reg.fit_transform(X)),c='b')
plt.xlabel("Area")
plt.ylabel("Price")
plt.show()
```
